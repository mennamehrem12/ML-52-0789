```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns, index=X_train.index)
X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns, index=X_test.index)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define base KNN model
knn = KNeighborsClassifier()

# Wrap in MultiOutputClassifier
multi_knn = MultiOutputClassifier(knn)

# Define parameter grid
param_grid = {
    'estimator__n_neighbors': [3,5,7, 9, 11, 13],
    'estimator__weights': ['uniform', 'distance'],
    'estimator__metric': ['minkowski', 'manhattan', 'chebyshev']
}

# Run GridSearchCV
grid_search = GridSearchCV(multi_knn, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train_scaled, y_train)

# Get best estimator and predict
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test_scaled)

# Convert predictions to DataFrame for evaluation
y_pred_df = pd.DataFrame(y_pred, columns=y_test.columns)

# Print classification reports for each output
for col in y_test.columns:
    print(f"\nClassification report for {col}:\n")
    print(classification_report(y_test[col], y_pred_df[col]))

    # Confusion Matrix
    cm = confusion_matrix(y_test[col], y_pred_df[col])
    print(f"Confusion Matrix for {col}:\n{cm}")

    # Plot confusion matrix as a heatmap
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['No', 'Yes'], yticklabels=['No', 'Yes'])
    plt.title(f'Confusion Matrix for {col}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

# Print best hyperparameters
print("Best hyperparameters found: ", grid_search.best_params_)
```

KNN achieved 66% accuracy for ADHD_Outcome and 63% accuracy for Sex_F. While the model performs decently on the majority class (ADHD-positive, recall: 93%), it struggles significantly with the minority class (ADHD-negative, recall: 7%). This suggests that KNN is heavily influenced by the dominant class, making it unreliable for detecting non-ADHD cases. For Sex_F, the model does better for class 0 (which is men) (recall: 87%) but performs poorly for class 1 (recall: 16%), reinforcing the class imbalance problem.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define base Decision Tree model
tree = DecisionTreeClassifier(random_state=42)

# Wrap in MultiOutputClassifier
multi_tree = MultiOutputClassifier(tree)

# Define hyperparameter grid
param_grid = {
    'estimator__max_depth': [3, 5, 10, None],
    'estimator__min_samples_split': [2, 5, 10],
    'estimator__min_samples_leaf': [1, 2, 4],
    'estimator__criterion': ['gini', 'entropy']
}

# Initialize and run GridSearchCV
grid_search = GridSearchCV(multi_tree, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train_scaled, y_train)

# Get best model and make predictions
best_tree_model = grid_search.best_estimator_
y_pred = best_tree_model.predict(X_test_scaled)

# Convert predictions to DataFrame for evaluation
y_pred_df = pd.DataFrame(y_pred, columns=y_test.columns)

# Print classification report for each target
for col in y_test.columns:
    print(f"\nClassification report for {col}:\n")
    print(classification_report(y_test[col], y_pred_df[col]))

    # Confusion Matrix
    cm = confusion_matrix(y_test[col], y_pred_df[col])
    print(f"Confusion Matrix for {col}:\n{cm}")

    # Plot confusion matrix as a heatmap
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['No', 'Yes'], yticklabels=['No', 'Yes'])
    plt.title(f'Confusion Matrix for {col}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

# Print best parameters
print("Best hyperparameters found: ", grid_search.best_params_)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.multioutput import MultiOutputClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix

# Load data
x1 = train_df
y1 = pd.read_excel('TRAINING_SOLUTIONS.xlsx')
y1 = y1.set_index('participant_id')

# Split the data
X_train, X_test, y_train, y_test = train_test_split(x1, y1, test_size=0.3, random_state=42)

# Scale the data
scaler = StandardScaler()
x1_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns, index=X_train.index)
x1_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns, index=X_test.index)

# Base model
model_log = LogisticRegression(penalty='l1', solver='liblinear', max_iter=1000)

# Wrap it with MultiOutputClassifier
multi_log = MultiOutputClassifier(model_log)

# Define parameter grid
param_grid = {
    'estimator__C': [0.01, 0.1, 1, 10, 100],  # Regularization strength
}

# Grid search with cross-validation
grid_search = GridSearchCV(multi_log, param_grid, cv=5, scoring='f1_macro', verbose=1, n_jobs=-1)
grid_search.fit(x1_train_scaled, y_train)

# Best estimator
best_model = grid_search.best_estimator_

# Predict
y_pred = best_model.predict(x1_test_scaled)

# Evaluate
for i, column in enumerate(y1.columns):
    print(f"\nClassification report for {column}:\n")
    print(classification_report(y_test.iloc[:, i], y_pred[:, i]))

    # Confusion Matrix
    cm = confusion_matrix(y_test.iloc[:, i], y_pred[:, i])
    print(f"Confusion Matrix for {column}:\n{cm}")

    # Plot confusion matrix
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['No', 'Yes'], yticklabels=['No', 'Yes'])
    plt.title(f'Confusion Matrix for {column}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

Logistic Regression provided the most balanced recalls (25% for ADHD-negative, 72% for ADHD-positive) but still had mediocre accuracy (58%) and it showed worse accuracy than the other models. Unlike tree-based models, it does not completely ignore the minority class, but its performance remains weak. For Sex_F, it shows a similar trend (68% recall for class 0, 31% for class 1). This suggests that while Logistic Regression is less biased than other models, it still suffers from class imbalance.

The results reveal a recurring issue; class imbalance, where models perform well on the majority class 
but poorly on the minority class. This is particularly evident in the ADHD_Outcome task, where class 1 (ADHD-positive) 
dominates, leading to high recall for this class but near-failure in detecting class 0 (ADHD-negative).
